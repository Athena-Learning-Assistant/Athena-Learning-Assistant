```python
import pandas as pd
from datasets import load_dataset

# Memuat dataset
dataset = load_dataset("indonlp/cendol_collection_v2")

# Mengambil train split
train_data = dataset['train']

# Memproses data untuk membuat kolom 'text' dengan format yang diinginkan
def format_text(row):
    return f"<s>[INST]{row['input']}[/INST]{row['output']}</s>"

# Menerapkan fungsi format_text pada setiap baris di train_data
formatted_data = train_data.map(lambda row: {'text': format_text(row)})

# Konversi ke DataFrame dengan kolom bernama 'text'
df = pd.DataFrame({'text': formatted_data['text']})

# Mengatur Pandas untuk menampilkan semua konten sel
pd.set_option('display.max_colwidth', None)

# Menampilkan baris pertama
print(df.iloc[0])
```

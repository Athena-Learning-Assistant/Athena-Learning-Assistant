```python
from google.colab import auth
auth.authenticate_user()

from google.cloud import storage
client = storage.Client()
```

```python
from huggingface_hub import snapshot_download

# Ganti dengan nama model Anda
model_name = "Equinox391/Athena-gguf-llama2"

# Download model dari Hugging Face
model_dir = snapshot_download(model_name)
```

```python
import os
from google.cloud import storage

bucket_name = "athena-dev-bucket"
bucket = client.bucket(bucket_name)

# Fungsi untuk mengunggah file
def upload_to_bucket(blob_name, file_path, bucket):
    blob = bucket.blob(blob_name)
    blob.upload_from_filename(file_path)
    print(f"File {file_path} diunggah ke {blob_name}.")

# Mengunggah semua file dalam direktori model
for root, dirs, files in os.walk(model_dir):
    for file in files:
        file_path = os.path.join(root, file)
        blob_name = os.path.relpath(file_path, model_dir)
        upload_to_bucket(blob_name, file_path, bucket)
```
